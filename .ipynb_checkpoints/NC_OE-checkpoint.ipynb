```python
import hail as hl
import re
from pyspark.sql import SparkSession
import os

config = {
    'spark.driver.memory': '30g',
    'spark.executor.memory': '300g',
    'spark.executor.cores': '3',
    'spark.driver.memoryOverhead': '40g'
}

hl.init(spark_conf=config, master='local[96]')
rg = hl.default_reference('GRCh38')
rg = hl.get_reference('GRCh38')
rg.add_sequence('references_Homo_sapiens_assembly38.fasta.gz', 'references_Homo_sapiens_assembly38.fasta.fai')

spark = SparkSession.builder.appName("Parquet Example").getOrCreate()
```

```python
#Create hail table of all bed positions of interest
bed_data = spark.read.parquet('features_by_position.parquet')
bed_ht = hl.Table.from_spark(bed_data)
bed_ht = bed_ht.annotate(variant_str = bed_ht.hg38_chr + ":" + hl.str(bed_ht.hg38_position))
bed_ht = bed_ht.annotate(locus = hl.parse_locus(bed_ht.variant_str,reference_genome='GRCh38'))
bed_ht = bed_ht.annotate(
    ref=bed_ht.locus.sequence_context()
)
bed_ht = bed_ht.annotate(
    possible_alts=hl.if_else(
        bed_ht.ref != "N",
        hl.set(['A', 'T', 'C', 'G']).remove(bed_ht.ref),
        hl.set(['N'])
    )
)
bed_ht = bed_ht.explode('possible_alts')
bed_ht = bed_ht.annotate(
    alleles=hl.array([bed_ht.ref, bed_ht.possible_alts])
)
bed_ht = bed_ht.select(
    hg38_chr = bed_ht.hg38_chr,
    hg38_position = bed_ht.hg38_position,
    REF = bed_ht.ref,
    ALT = bed_ht.possible_alts
)
bed_ht.export('127vars.tsv.bgz')
```

```python
#Create Roulette Hail Table
recode = {f"{i}":f"chr{i}" for i in (list(range(1, 23)) + ['X', 'Y'])}
roulette_mt = hl.import_vcf('/home/ubuntu/Annotation/data/downloads/*_rate_v5.2_TFBS_correction_all.vcf.bgz',reference_genome='GRCh38', contig_recoding=recode)
roulette_mt = roulette_mt.rows()

roulette_mt = roulette_mt.annotate(
    roulette_AR_MR = hl.if_else(
        hl.is_defined(roulette_mt.info.AR),
        roulette_mt.info.AR,
        roulette_mt.info.MR
    )
)

roulette_mt = roulette_mt.select(
    qual = roulette_mt.qual,
    filters = roulette_mt.filters,
    roulette_AR_MR = roulette_mt.roulette_AR_MR
)

roulette_mt = roulette_mt.checkpoint('roulette_mt.ht', overwrite=True)

roulette_mt.count()
```

```python
primate_vars_file = spark.read.parquet('processed_counts_20241029.parquet')
primate_vars_ht = hl.Table.from_spark(primate_vars_file)
primate_vars_ht = primate_vars_ht.repartition(192)

primate_vars_ht = primate_vars_ht.annotate(variant_str = primate_vars_ht.hg38_chr + ":" + hl.str(primate_vars_ht.hg38_position))
primate_vars_ht = primate_vars_ht.select(
    locus = hl.parse_locus(primate_vars_ht.variant_str,reference_genome='GRCh38'), 
    region = primate_vars_ht.region,
    variant_binary = primate_vars_ht.variant_binary,
    null_percent = primate_vars_ht.null_percent
)

primate_vars_ht = primate_vars_ht.annotate(
    ref=primate_vars_ht.locus.sequence_context()
)

primate_vars_ht = primate_vars_ht.annotate(
    possible_alts=hl.if_else(
        primate_vars_ht.ref != "N",
        hl.set(['A', 'T', 'C', 'G']).remove(primate_vars_ht.ref),
        hl.set(['N'])
    )
)

primate_vars_ht = primate_vars_ht.explode('possible_alts')
#Note there are some positions with N ref which will have 3 possible alts

primate_vars_ht = primate_vars_ht.annotate(
    alleles=hl.array([primate_vars_ht.ref, primate_vars_ht.possible_alts])
)

primate_vars_ht = primate_vars_ht.select(
    locus = primate_vars_ht.locus, 
    alleles = primate_vars_ht.alleles,
    region = primate_vars_ht.region,
    primate_variant_binary = primate_vars_ht.variant_binary,
    primate_null_percent = primate_vars_ht.null_percent
)

primate_vars_ht = primate_vars_ht.key_by(primate_vars_ht.locus, primate_vars_ht.alleles)

roulette_mt = hl.read_table('roulette_mt.ht')

primate_vars_ht = primate_vars_ht.annotate(
    roulette_AR_MR = roulette_mt[primate_vars_ht.key].roulette_AR_MR,
    roulette_qual = roulette_mt[primate_vars_ht.key].qual, 
    roulette_filter = roulette_mt[primate_vars_ht.key].filters,
)

syn_sites = hl.import_table('/home/ubuntu/Annotation/data/downloads/all_hq_synonymous_variants.tsv.bgz')
syn_sites = syn_sites.annotate(variant_str = "chr" + syn_sites.CHROM + ":" + syn_sites.POS)
syn_sites = syn_sites.select(locus = hl.parse_locus(syn_sites.variant_str,reference_genome='GRCh38'), alleles = [syn_sites.REF, syn_sites.ALT], roulette_hq_syn = True)

syn_sites = syn_sites.key_by(syn_sites.locus, syn_sites.alleles)

primate_vars_ht = primate_vars_ht.annotate(roulette_hq_syn = syn_sites[primate_vars_ht.key].roulette_hq_syn)

primate_vars_ht = primate_vars_ht.checkpoint('primate_vars.ht', overwrite=True)

primate_vars_ht.count() #382823782
```

```python
from scipy.optimize import minimize
import numpy as np

primate_vars_ht = hl.read_table('primate_vars.ht')

scaling = primate_vars_ht.filter(primate_vars_ht.roulette_hq_syn == True)

scaling = scaling.key_by()

scaling = scaling.select(
    primate_variant_binary = scaling.primate_variant_binary,
    roulette_AR_MR = scaling.roulette_AR_MR
)

scaling = (scaling.group_by(scaling.roulette_AR_MR)
                .aggregate(
                    synObs = hl.agg.sum(scaling.primate_variant_binary),
                    synTotal = hl.agg.count()
                )
          )

obs_by_mu = scaling.to_pandas()

def poisson_function(x, obs_by_mu):
    return abs(sum((1 - np.exp(-1 * x * obs_by_mu["roulette_AR_MR"]))*obs_by_mu["synTotal"]) - sum(obs_by_mu["synObs"]))
    
x0 = np.array([1])
res = minimize(poisson_function, x0, args = (obs_by_mu), method='Nelder-Mead', options={'xatol': 1e-3, 'disp': False})

# this is the proper scaling factor for the mutation rate
k = res.x[0]

print(k)
```

```python
obs_by_mu['Ratio'] = obs_by_mu['synObs']/obs_by_mu['synTotal']
obs_by_mu.to_csv('bins.tsv', sep='\t')
```

```python
primate_vars_ht = hl.read_table('primate_vars.ht')

primate_vars_ht = primate_vars_ht.annotate(
    regions = primate_vars_ht.region.split(','),
    scaled_roulette_prob = (1 - hl.exp((-1) * k * primate_vars_ht.roulette_AR_MR))
).drop('region')

primate_vars_ht = primate_vars_ht.checkpoint('primate_vars_scaled.ht', overwrite=True)
```

```python
# First, filter the transcript consequences
primate_vars_ht = hl.read_table("primate_vars_vep.ht")

primate_vars_ht = primate_vars_ht.explode('regions')

primate_vars_ht = primate_vars_ht.annotate(
    transcript = primate_vars_ht.regions.split('_')[0],
    region_type = primate_vars_ht.regions.split('_')[1]
).drop('regions')

primate_vars_ht = primate_vars_ht.annotate(
    filtered_consequences = primate_vars_ht.vep.transcript_consequences.filter(
        lambda x: x.transcript_id == primate_vars_ht.transcript
    )
)

# Then extract the consequence terms from the filtered consequences
primate_vars_ht = primate_vars_ht.annotate(
    consequence_terms = hl.if_else(
        hl.len(primate_vars_ht.filtered_consequences) > 0,
        primate_vars_ht.filtered_consequences[0].consequence_terms,
        hl.empty_array(hl.tstr)
    )
)


# Finally clean up by dropping unnecessary fields
primate_vars_ht = primate_vars_ht.drop('vep', 'vep_proc_id', 'filtered_consequences', 'roulette_AR_MR')
primate_vars_ht = primate_vars_ht.checkpoint("primate_vars_consequence.ht")
```

```python
#Clinvar variants from 12/10/24
recode = {str(i): f"chr{i}" for i in list(range(1, 23)) + ['X', 'Y']}
recode['MT'] = 'chrM'
clinvar_data = hl.import_vcf('clinvar.vcf.bgz', reference_genome='GRCh38', contig_recoding=recode, skip_invalid_loci=True).rows()

hq = hl.set([["criteria_provided","_multiple_submitters","_no_conflicts"], ["practice_guideline"], ["reviewed_by_expert_panel"]])
clinvar_data = clinvar_data.filter(hq.contains(clinvar_data.info.CLNREVSTAT))

clinvar_data = clinvar_data.filter(clinvar_data.info.CLNVC == "single_nucleotide_variant")
#options are "Deletion","Duplication","Indel","Insertion","Inversion","Microsatellite","Variation","single_nucleotide_variant"

clinvar_data = clinvar_data.filter(clinvar_data.info.ORIGIN.contains('1'))
#many options here but a vast majority of variants are germline (1)

clinvar_data = clinvar_data.filter(clinvar_data.info.CLNSIG.length() == 1)
#removes 4 variants with _low_penetrance

clinvar_data = clinvar_data.select(CLNSIG = clinvar_data.info.CLNSIG[0])

clinvar_data = clinvar_data.checkpoint('filtered_clinvar.ht')
```
